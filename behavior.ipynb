# Treating null values
### Size

```python
behavior['Size'].unique()
```

```python
behavior.groupby('Category')['Size'].unique()
```

```python
behavior.groupby('Item Purchased')['Size'].unique()
```

```python
# treating null values or replacing all thevalues in the electronics cateogry
behavior.loc[df['Category']=='Electronics','Size']="Not applicable"
#df['Category']=='Electronics':-Ye condition check karta hai ki kaunsi rows me category Electronics hai.
# df.loc[ condition , column ]:-.loc[] ka use specific rows aur columns select karne ke liye hota hai.
#df.loc[df['Category']=='Electronics','Size']:-jin rows me Electronics hai,unka Size column select karo
#"Not applicable":-Ab selected rows ke Size column me ye value daali ja rahi hai
```

```python
behavior.groupby('Category')['Size'].unique()
```

```python
# treating null values or replacing all thevalues in the Accessories cateogry
behavior.loc[behavior['Category']=='Accessories','Size']="Free size"
```

```python
behavior.groupby('Category')['Size'].unique()
```

```python
# treating null values or replacing all thevalues in the Footwear  cateogry
behavior.loc[behavior['Category']=='Footwear','Size']="Not available"
```

```python
behavior.groupby('Category')['Size'].unique()
```

```python
# treating null values or replacing all the values in the clothing cateogry
behavior.loc[(behavior['Category']=='Clothing')&(behavior['Size'].isnull()),'Size']=behavior[behavior['Category']=='Clothing']['Size'].mode()[0]
# Purpose kya hai?:-Clothing category ke andar jaha Size missing (null) hai,waha sabse zyada aane wala size fill kar do.
#df['Cateogry']=='Clothing':-Kaunsi rows Clothing category ki hain?
#df['Size'].isnull():-Ye batata hai Size column me kaha missing values hain.
#(df['Cateogry']=='Clothing') & (df['Size'].isnull()):-Dono conditions combine karna
#df.loc[condition, 'Size']:-Ye selected rows ke Size column ko target karta hai.
#df[df['Cateogry']=='Clothing']['Size'].mode()[0]:-iska matlb niche hai
# df[df['Cateogry']=='Clothing']:-Sirf Clothing rows le lo.
#['Size']:-Sirf Size column le lo.
#.mode():-Most frequent value nikalo

```

# Review rating

```python
behavior.head(5)
```

```python
behavior.isnull().sum()
```

```python
# mean of review rating
behavior['Review Rating'].mean()
```

```python
#median of review rating
behavior['Review Rating'].median()
```

```python
#mode of review rating
behavior['Review Rating'].mode()
```

```python
#item ka average review rating
behavior.groupby("Item Purchased")["Review Rating"].mean()
```

```python
# filling null values of review rating with the mean of each item purchased
behavior["Review Rating"]=behavior["Review Rating"].fillna(behavior.groupby("Item Purchased")["Review Rating"].transform('mean'))
```

```python
behavior.isnull().sum()
```

```python
behavior.head(10)
```

```python
# so review rating me zero kaphi jayada bar aa rha hai to usko 2 place tk krne ke liye hum round off ka use krenge
behavior["Review Rating"]=behavior["Review Rating"].round(2)
```

```python
behavior.head(10)
```

# Previous Purchases :-

```python
# filling previous purchase with zero beacause agr ek se jyada bar purchase krta to itna null value nhi hota 
behavior["Previous Purchases"]=behavior["Previous Purchases"].fillna(0)
```

# Purchase Amount (USD)

```python
# average price of the item 
behavior.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
# filling null values of Purchase Amount (USD) with the average price of the item
behavior["Purchase Amount (USD)"]=behavior["Purchase Amount (USD)"].fillna(behavior.groupby("Item Purchased")["Purchase Amount (USD)"].transform('mean'))
```

```python
behavior.isnull().sum()
```

```python
#Ab sirf Category column me null values bachi hain.Ye null values isliye aayi kyunki kuch items correct_mapping me match nahi hue the.Ab unko fix karne ke liye pehle dekho kaunse items ki wajah se null aa raha hai.
behavior[behavior['Category'].isnull()]['Item Purchased'].unique()
#Isse pata chalega kaunse items unmapped hain.
```

```python
# now gloves item mapped nhi hai isliye category me null value aa rha tha
correct_mapping['Gloves'] = 'Accessories'
behavior['Category'] = behavior['Item Purchased'].map(correct_mapping)
```
